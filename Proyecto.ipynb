## La transformada discreta de Fourier 
Permite realizar el análisis espectral de una señal. Nos lleva del espacio de tiempo discreto al espacio de frecuencias, donde podemos obtener información sobre las componentes frecuenciales predominantes de una señal.

##  Identificación Frecucencia Predominante en Señal Discreta

Se deben crear señales que estén compuestas por la frecuencia promedio del sonido de las vocales. </br>
| Vocal 	| Formante 1 	| Formante 2 	|</br>
|-------	|------------	|------------	|</br>
| a     	| 1000 Hz    	| 1400 Hz    	|</br>
| e     	| 500 Hz     	| 2300 Hz    	|</br>
| i     	| 320 Hz     	| 3200 Hz    	|</br>
| o     	| 500 Hz     	| 1000 Hz    	|</br>
| u     	| 320 Hz     	| 800 Hz     	|


```python
from IPython.display import Image  #Para poder mostra imagenes
import numpy as np                 #Para el manejos de los conjuntos de datos             
import scipy.fftpack as fourier    #Para el calculo de FFT
import matplotlib.pyplot as plt    #Para mostrar las gráficas
import scipy.io.wavfile as waves   #Para poder leer archivos .WAV
import winsound                    #Para reproducir sonidos
```

# Vocal a

```python
TiempoMuestreo = 0.001                     # Definimos un tiempo y frecuencia de muestreo
FrecuenciaMuestreo=1/TiempoMuestreo
a1 = 2*np.pi*1000                          # Definimos una frecuencia de 1000 Hz para la señal a1
a2 = 2*np.pi*1400                          # Definimos una frecuencia de 1400 Hz para la señal a2
n = TiempoMuestreo*np.arange(0, 1000)
ruido = np.random.random(len(n))
x = 3*np.sin(a1*n)+2.3*np.sin(a2*n)+ruido  # Construimos una señal compuesta por la señal a1 y a2, y agregamos ruido
plt.plot(n,x,'.-')
plt.xlabel('Tiempo (s)', fontsize='14')
plt.ylabel('Amplitud', fontsize='14')
plt.show()
```

```python
%matplotlib inline

gk = fourier.fft(x)                             # Calculamos la FFT
M_gk = abs(gk)                                  # Calculamos la Magnitud de la FFT

F = FrecuenciaMuestreo*np.arange(0, len(x))/len(x)              # Definimos el Vector de Frecuencias

plt.plot(F, M_gk)
plt.xlabel('Frecuencia (Hz)', fontsize='14')
plt.ylabel('Amplitud FFT', fontsize='13')
plt.show()
```
